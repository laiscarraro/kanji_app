```python
import pandas as pd

subs = pd.read_csv('data/subtitles.csv', sep=';')
subs.head()
```

```python
import re

patterns = [
    's\d+e\d+',
    '\d+.srt',
    '\\b\d+\\b'
]
pattern = '|'.join(patterns)

subs['episode'] = subs.filename.apply(
    lambda filename:
        re.sub(
            '.srt', '',
            ''.join(
                re.findall(pattern, filename.lower())
            )
        )
)

subs
```
