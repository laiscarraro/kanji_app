```python
from ankipandas import Collection
import pandas as pd

col = Collection()
metadata = col.cards
metadata.index = metadata.nid
data = col.notes

cards = pd.merge(metadata, data, left_index=True, right_index=True)
cards.head()
```

```python
import re
kanji_decks = ['RRTK Recognition Remembering The Kanji v2']
words_decks = ['日本語 JLPT Tango N5 MIA Japanese', '日本語 JLPT Tango N4 Omega']

kanji_cards = cards[cards.cdeck.apply(lambda a: re.sub('\x1f', ' ', a)).isin(kanji_decks)]
words_cards = cards[cards.cdeck.apply(lambda a: re.sub('\x1f', ' ', a)).isin(words_decks)]
```

```python
kanji_cards['kanji'] = kanji_cards['nflds'].apply(lambda a: a[0])
words_cards['word'] = words_cards['nflds'].apply(lambda a: a[0])
```

```python
unlocked_kanji = kanji_cards[kanji_cards['cqueue'] != 'new'][['kanji', 'ctype', 'cqueue', 'cdeck']]
unlocked_words = words_cards[words_cards['cqueue'] != 'new'][['word', 'ctype', 'cqueue', 'cdeck']]
```
